```python
cap = cv2.VideoCapture('input_video.mp4')
fps = cap.get(cv2.CAP_PROP_FPS)
frame_count = cap.get(cv2.CAP_PROP_FRAME_COUNT)
frame_width = cap.get(cv2.CAP_PROP_FRAME_WIDTH)
frame_height = cap.get(cv2.CAP_PROP_FRAME_HEIGHT)

videoWriter4 = cv2.VideoWriter('output_video4.avi',cv2.VideoWriter_fourcc('M','J','P','G'), 24, (int(cap.get(cv2.CAP_PROP_FRAME_WIDTH)), int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))))

for i in range(int(frame_count)):
    ret, frame = cap.read()
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    # Apply Canny edge detection
    edges = cv2.Canny(gray, threshold1=50, threshold2=150)
    # Convert the edges to color for visualization
    edges_color = cv2.cvtColor(edges, cv2.COLOR_GRAY2BGR)
    # Combine the original frame and the edge image using bitwise OR
    output_frame = cv2.bitwise_or(frame, edges_color)
    # Write the output frame to the output video
    videoWriter4.write(output_frame)

videoWriter4.release()
cap.release()
```

```python
input_file = 'input_video.mp4'
watermark_file = 'watermark_video.mp4'

cap_input = cv2.VideoCapture(input_file)
cap_watermark = cv2.VideoCapture(watermark_file)

fps = cap_input.get(cv2.CAP_PROP_FPS)
frame_width = int(cap_input.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap_input.get(cv2.CAP_PROP_FRAME_HEIGHT))

videoWriter5 = cv2.VideoWriter('output_video5.avi',cv2.VideoWriter_fourcc('M','J','P','G'), 24, (frame_width, frame_height))

while cap_input.isOpened() and cap_watermark.isOpened():
    ret_input, frame_input = cap_input.read()
    ret_watermark, frame_watermark = cap_watermark.read()

    if not ret_input or not ret_watermark:
        break

    # Resize the watermark frame to a desired size
    frame_watermark = cv2.resize(frame_watermark, (frame_width, frame_height))

    # Create a mask from the watermark frame (use one of the color channels)
    mask = frame_watermark[:, :, 0]

    # Invert the mask
    mask_inv = cv2.bitwise_not(mask)

    # Convert the mask to 3 channels for proper blending
    mask_rgb = cv2.cvtColor(mask_inv, cv2.COLOR_GRAY2BGR)

    # Blend the input frame and watermark frame using the mask
    blended_frame = cv2.bitwise_and(frame_input, mask_rgb) + cv2.bitwise_and(frame_watermark, frame_watermark)

    # Write the blended frame to the output video
    videoWriter5.write(blended_frame)

cap_input.release()
cap_watermark.release()
videoWriter5.release()

```

```python
import cv2
import numpy as np

input_file = 'videoplayback.mp4'
cap = cv2.VideoCapture(input_file)

fps = cap.get(cv2.CAP_PROP_FPS)
frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

videoWriter6 = cv2.VideoWriter('output_video6.avi',cv2.VideoWriter_fourcc('M','J','P','G'), 24,  (frame_width, frame_height))

mask = np.zeros((frame_height, frame_width, 3), dtype=np.uint8)

lk_params = dict(winSize=(15, 15), maxLevel=2, criteria=(cv2.TERM_CRITERIA_EPS | cv2.TERM_CRITERIA_COUNT, 10, 0.03))

ret, prev_frame = cap.read()
prev_gray = cv2.cvtColor(prev_frame, cv2.COLOR_BGR2GRAY)

while cap.isOpened():
    ret, frame = cap.read()

    if not ret:
        break

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    flow = cv2.calcOpticalFlowFarneback(prev_gray, gray, None, 0.5, 3, 15, 3, 5, 1.2, 0)

    magnitude, angle = cv2.cartToPolar(flow[..., 0], flow[..., 1])
    magnitude[np.isnan(magnitude)] = 0
    magnitude = np.clip(magnitude, 0, 255)
    
    mask[..., 0] = angle * 180 / np.pi / 2
    mask[..., 1] = 255  # Set the saturation value to maximum
    mask[..., 2] = cv2.normalize(magnitude, None, 0, 255, cv2.NORM_MINMAX)

    mask_bgr = cv2.cvtColor(mask, cv2.COLOR_HSV2BGR)

    # Overlay the frame and the mask using addition
    output_frame = cv2.add(frame, mask_bgr)

    videoWriter6.write(output_frame)

    prev_frame = frame
    prev_gray = gray

cap.release()
videoWriter6.release()
```

```python
import cv2
import numpy as np


cap = cv2.VideoCapture("videoplayback.mp4")
ret, frame1 = cap.read()
prvs = cv2.cvtColor(frame1, cv2.COLOR_BGR2GRAY)

# Define the output video settings
output_video = cv2.VideoWriter('output_video.mp4',
                               cv2.VideoWriter_fourcc(*'mp4v'),
                               cap.get(cv2.CAP_PROP_FPS),
                               (frame1.shape[1], frame1.shape[0]))

while True:
    ret, frame2 = cap.read()
    if not ret:
        break

    next_frame = cv2.cvtColor(frame2, cv2.COLOR_BGR2GRAY)

    flow = cv2.calcOpticalFlowFarneback(prvs, next_frame, None, 0.5, 3, 15, 3, 5, 1.2, 0)

    # Compute the magnitude and angle of the optical flow vectors
    mag, ang = cv2.cartToPolar(flow[..., 0], flow[..., 1])
    # Handle NaN or Inf values in mag array
    if np.isnan(mag).any() or np.isinf(mag).any():
        mag = np.nan_to_num(mag)
    # Create a mask image to draw the motion vectors
    mask = np.zeros_like(frame2)

    # Set the motion vector display parameters
    arrow_scale = 5
    arrow_color = (0, 255, 0)  # Green color

    # Draw the motion vectors as arrows on the mask image
    for y in range(0, frame2.shape[0], 10):
        for x in range(0, frame2.shape[1], 10):
            dx = int(flow[y, x, 0])
            dy = int(flow[y, x, 1])
            cv2.arrowedLine(mask, (x, y), (x + dx, y + dy), arrow_color, thickness=1, tipLength=0.5)

    # Combine the original frame with the motion vectors mask
    output_frame = cv2.add(frame2, mask)

    output_video.write(output_frame)

    prvs = next_frame

cap.release()
output_video.release()
cv2.destroyAllWindows()
```
